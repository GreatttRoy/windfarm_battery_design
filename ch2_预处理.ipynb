```python
import pandas as pd
import numpy as np
from datetime import datetime, timezone
```

```python
df_raw = pd.read_excel('data/ods031.xlsx')
df_raw
```

```python
def remove_timezone(dt):
    # HERE `dt` is a python datetime
    # object that used .replace() method
    return dt.replace(tzinfo=None)
```

```python
df_raw['Datetime'] = pd.to_datetime(df_raw['Datetime'])
df_raw['Datetime'] = df_raw['Datetime'].apply(remove_timezone)

df_raw = df_raw.sort_values(by='Datetime')
df_raw = df_raw.reset_index(drop=True)
```

```python
df_anly = df_raw.copy()
df_anly = df_anly[df_anly['Region'] =='Federal']
df_anly = df_anly.sort_values(by='Datetime',ascending = True)
df_anly = df_anly.reset_index(drop=True)
df_anly.iloc[28996:29004]
```

```python
num = 9
df_slim_dead = df_slim_cal
# 假设 df_slim_cal 是您的DataFrame，且已有 'Measured & Upscaled' 这一列

# 创建一个新列以标记值何时改变
df_slim_dead['value_changed'] = df_slim_dead['Measured & Upscaled'].shift() != df_slim_dead['Measured & Upscaled']

# 为每个连续段分配组ID
df_slim_dead['group_id'] = df_slim_dead['value_changed'].cumsum()

# 计算每个组的大小
group_sizes = df_slim_dead.groupby('group_id').size()

# 筛选出大小大于等于6的组ID
valid_group_ids = group_sizes[group_sizes >= num].index

# 根据筛选出的组ID，选择原DataFrame中的行
filtered_df = df_slim_cal[df_slim_cal['group_id'].isin(valid_group_ids)]

# 如果需要，可以从 filtered_df 中删除辅助列
filtered_df = filtered_df.drop(['value_changed', 'group_id'], axis=1)

# 此时，filtered_df 包含了满足条件的行
filtered_df.to_csv('ch2fig/死数数据集.csv')

# 继续使用上面创建的 df_slim_dead 和 group_sizes

# 使用 groupby 和 agg 来计算每个组的统计信息
stats = df_slim_dead.groupby('group_id').agg(
    unique_value=pd.NamedAgg(column='Measured & Upscaled', aggfunc='first'),  # 假设连续段的值是相同的，取第一个即可
    length=pd.NamedAgg(column='Measured & Upscaled', aggfunc='size'),
    start_time=pd.NamedAgg(column='Datetime', aggfunc='min'),
    end_time=pd.NamedAgg(column='Datetime', aggfunc='max')
)

# 筛选出长度大于等于4的记录
filtered_stats = stats[stats['length'] >= num]

# 保存统计信息到CSV文件
filtered_stats.to_csv('ch2fig/连续数字统计信息.csv')

filtered_stats
```
