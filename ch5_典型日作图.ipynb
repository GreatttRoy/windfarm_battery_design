# 夏季典型日画图


```python
# 第一个典型日

df_summer_method1 = pd.read_csv('ch5_data/analyse_by_date/2022-06-19/2022-06-19_abs_30_anly.csv')
df_summer_method2 = pd.read_csv('ch5_data/analyse_by_date/2022-06-19/2022-06-19_range_1_anly.csv')
df_summer_method3 = pd.read_csv('ch5_data/analyse_by_date/2022-06-19/2022-06-19_range_30_anly.csv')
df_summer_method4 = pd.read_csv('ch5_data/analyse_by_date/2022-06-19/2022-06-19_range_with_E0.5_30_anly.csv')
df_summer_method1
```

## 有功功率图

### 方案1 VS 方案3

```python

data = df_summer_method1.copy()

# 使用Plotly创建交互式图表
fig = make_subplots(rows=1, cols=1) # 实际上只画了一个图
# 添加数据线
fig.add_trace(go.Scatter(x=data['Datetime'], y=df_summer_method1['P_W_list'], mode='lines+markers', name='原有功功率'))
fig.add_trace(go.Scatter(x=data['Datetime'], y=df_summer_method1['P_CW'], mode='lines+markers', name='方案1有功功率'))
fig.add_trace(go.Scatter(x=data['Datetime'], y=df_summer_method3['P_CW'], mode='lines+markers', name='方案3有功功率'))


# 设置图表布局
fig.update_layout(
    xaxis_title='日期时间',
    yaxis_title='功率值 (MW)',
    legend_title='有功功率功率值',
    font=dict(family="SimHei", size=20)
)

# 设置X轴为日期时间格式
fig.update_xaxes(tickformat="%Y-%m-%d %H:%M:%S")
fig.show()
```

## 波动图

### 方案1 VS 方案3

```python

data = df_summer_method1.copy()

# 使用Plotly创建交互式图表
fig = make_subplots(rows=1, cols=1) # 实际上只画了一个图


# 添加数据线
fig.add_trace(go.Scatter(x=data['Datetime'], y=df_summer_method1['original_flucuation'], mode='lines+markers', name='原功率变化差'))
fig.add_trace(go.Scatter(x=data['Datetime'], y=df_summer_method1['after_flucuation'], mode='lines+markers', name='方案1功率变化差'))
fig.add_trace(go.Scatter(x=data['Datetime'], y=df_summer_method3['after_flucuation'], mode='lines+markers', name='方案3功率变化差'))
fig.add_trace(go.Scatter(x=data['Datetime'], y=df_summer_method1['E'], mode='lines+markers', name='方案1能量'))
fig.add_trace(go.Scatter(x=data['Datetime'], y=df_summer_method3['E'], mode='lines+markers', name='方案3能量'))

# 添加基线
fig.add_hline(y=upper_baseline, line_dash="dash", line_color="orange", annotation_text="上波动考核线", annotation_position="top right")
fig.add_hline(y=lower_baseline, line_dash="dash", line_color="red", annotation_text="下波动考核线", annotation_position="bottom right")

# 设置图表布局
fig.update_layout(
    xaxis_title='日期时间',
    yaxis_title='功率变化差 (MW)或能量 (MW.h)',
    legend_title='功率变化差或能量',
    font=dict(family="SimHei", size=20)
)

# 设置X轴为日期时间格式
fig.update_xaxes(tickformat="%Y-%m-%d %H:%M:%S")
fig.show()
```
